```python
#importing libraries
import cv2
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
#loading the image
image=cv2.imread('pizza.jpg')
```

**Reading the Image**

```python
im = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,7))
plt.imshow(im)
plt.title("ORIGINAL IMAGE")
plt.axis("off")
plt.show()
```

**Converting the Original Image to GrayScale**

```python
img_gray = cv2.cvtColor(im , cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(10,6))
plt.imshow(img_gray,cmap="gray")
plt.title("GRAYSCALE IMAGE")
plt.axis("off")
plt.show()
```

**Inverting the New Image**

```python
img_invert = cv2.bitwise_not(img_gray)
plt.figure(figsize=(10,6))
plt.imshow(img_invert,cmap="gray")
plt.title("INVERTED IMAGE")
plt.axis("off")
plt.show()
```

 **Blurring the Image Using Gaussian Function**

```python
img_blur = cv2.GaussianBlur(img_invert,(21,21),0)
plt.figure(figsize=(10,6))
plt.imshow(img_blur,cmap="gray")
plt.title("BLURRED IMAGE")
plt.axis("off")
plt.show()
```

**Converting Image to Pencil Image**

```python
img_pencil = cv2.divide(img_gray,255-img_blur,scale=255)
plt.figure(figsize=(10,6))
plt.imshow(img_pencil,cmap="gray")
plt.title("FINAL IMAGE")
plt.axis("off")
plt.show()
```
